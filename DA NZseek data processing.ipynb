```python
def update_time(post_data):
    time = post_data[0]
    if 'ago' in time:
        (digit, unit) = re.findall(r"^(\d*)(\w*).? ago",time)[0]   
        digit = int(digit)
        
        if 'd' in unit:
            new_time = digit * 24
        if 'm' in unit:
            new_time = digit // 60
        if 'h' in unit:
            new_time = digit
            
        time = str(new_time) + 'h ago'

    return time
```

```python
new_time = time.apply(lambda x: update_time(x))
```

```python
def update_address(add):
    if 'area' in add:
        loc, area =  re.findall(r"^location: (.*)area: (.*)", add)[0]
        loc = remove_duplicate(loc)
        area = remove_duplicate(area)
        print(loc, area)
    else:
        loc =  re.findall(r"^location: (.*)", add)[0]
        loc = remove_duplicate(loc)
        print(loc)
```

```python
def remove_duplicate(s):
    string_length = len(s)
    
    start = string_length // 2
    for i in range(start):
        first_string = s[ : start - i]
        second_string = s[start - i : 2 * (start - i)]
        
        if first_string == second_string:
            return first_string
    return s
```

```python
fix_address = address.apply(lambda x: update_address(x))
```
